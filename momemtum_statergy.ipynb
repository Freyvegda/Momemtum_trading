```python
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

```python
import yfinance as yf

symbol = "GLD"
gld = yf.download(symbol)
day=np.arange(1,len(gld)+1)
gld['day']=day
gld.drop(columns=['Adj Close', 'Volume'], inplace= True)
gld=gld[['day', 'Open', 'High', 'Low', 'Close']]
gld.head()
```

```python
gld.info()
```

Adding moving averages to the df

```python
gld['9_day']= gld['Close'].rolling(9).mean().shift()
gld['21_day']= gld['Close'].rolling(21).mean().shift()
gld[20:25]
```

adding signal

```python
#adding signal to show if to go long or short
#long signal
gld['Signal'] = np.where(gld['9_day']>gld['21_day'], 1, 0)
#short signal
gld['Signal'] = np.where(gld['9_day']<gld['21_day'], -1, gld['Signal'])
#droping the 21 days value
gld.dropna(inplace=True)
gld.head()
```

Adding return column

```python
gld['Return']=np.log(gld['Close']).diff()
gld['System_return']= gld['Signal']*gld['Return']
gld['Entry']= gld.Signal.diff()
gld.head()
```

ploting the lines

```python
plt.rcParams['figure.figsize'] = 12, 6
plt.grid(True, alpha = .3)
plt.plot(gld.iloc[-252:]['Close'], label = 'GLD')
plt.plot(gld.iloc[-252:]['9_day'], label = '9_day')
plt.plot(gld.iloc[-252:]['21_day'], label = '21_day')
plt.plot(gld[-252:].loc[gld.Entry == 2].index, gld[-252:]['9_day'][gld.Entry == 2], '^',
         color = 'g', markersize = 12)
plt.plot(gld[-252:].loc[gld.Entry == -2].index, gld[-252:]['21_day'][gld.Entry == -2], 'v',
         color = 'r', markersize = 12)
plt.legend(loc=2);
```

```python
plt.plot(np.exp(gld['Return']).cumprod(), label='Buy/Hold')
plt.plot(np.exp(gld['System_return']).cumprod(), label='System')
plt.legend(loc=2)
plt.grid(True, alpha=.3)
```
